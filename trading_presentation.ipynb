## Explore Indicators Based on the Factors Above

```python
def run_backtest(macd_m1, macd_m2, macd_enter, macd_exit, rsi_upper, rsi_lower, start, start_trading, stop_trading, indicator, slip, y):
    macd_signal_df = ms.macd_rsi_signals(indicator, y['SI1:RSI'], macd_m1, macd_m2, macd_enter, macd_exit, rsi_upper, rsi_lower,
                    start_trading, stop_trading, plot = False, plot_start = start)
    # Backtest Inputs
    slippage_max = slip['WindowMax']
    slippage_min = slip['WindowMin']

    backTest = BT(INIT_CAP, y['SI1:close'], macd_signal_df, STOPLOSS, LOT_SIZE, COSTS, slippage_min, slippage_max)
    backTest.run_backtest()
    return backTest
```

## Optimizing Trading Size use Kelly Criterion

Consider a situation that when we win, we win $w$ percent, when we lose we $l$ percent. Each bet is a fraction $f$ of our total bankroll. So for $n$ wins and $m$ losses, the gain factor is 
$$
\mathcal{G}(f) = \left(1+fw\right)^n\left(1-fl\right)^m
$$
Or per trade we have 
$$
\mathcal{G}(f)^{1/(n+m)} = \left(1+fw\right)^p\left(1-fl\right)^{1-p}
$$
where $p = \frac{n}{n+m}$. Using the fact that $\mathcal{G}$ is concave, taking the log of $\mathcal{G}$ and derivative with respect to $f$ equal to zero, leads to optimal $f$. 
$$
f^* = \frac{pw - (1-p)l}{wl}
$$
We can estimate the win probability $p$ using Baye's Theorem where $\pi(p)$ is a prior distribution of $p$. (In this case uniform distribution is enough).

$$
\theta(p|w,N) = \frac{\pi(p)P(w|p,N)}{\int_0^1 \pi(p)P(w|p,N)dp}
$$
where $P(w|p,N)$ is the probability distribution of $w$ given $N$ and $p$. $P$ is binomial
$$
P(w|p,N) = \begin{pmatrix} N \\ w \end{pmatrix}p^w(1-p)^{N-w}.
$$
This leads to 
$$
\theta(p|w,N) =\frac{(N+1)!}{w!(N-w)!} p^w(1-p)^{N-w}.
$$
with mean 
$$
\hat{p} = \frac{w+1}{N+2}
$$

```python
LOT_SIZE = anal.compute_kelly_fraction(ret)
print("New sizing size of our strategy based on Kelly Criterion : {:.2f}%".format(LOT_SIZE * 100))
```

Interesting enough, our initial guess of 2% wasn't too far from the "optimal" ratio according to the Kelly Criterion.

# Part b - Risk Management & Drawdown Measurement

## Lot Size

Each trade will be 2% of the total account size as this is most typical. Further, this helps ensure the size of the order is filled at the price we want and not allowing the market to move before we are fully filled. Note that we can adjust this in the backtester to see how results change. However, we will not be able to see market impact from this backtester so it is safe to assume the typical lot size for this.

## Shutdown Rule

The strategy has a shutdown rule of 10% of the traded amount. However, the training backtest results show that our stop loss was never triggered. Further the the 95% Cornish-Fischer VaR and 95% Historic VaR were 1% and 1.5% respectively. Therefore, we can reduce the stop loss to less than 10%, however given the mean-reversion of our signal, it may be better to allow the trade to continue vs. stopping out and taking the loss. Lastly, the max drawdown was about 20%, which was lower than the market over the same time period
